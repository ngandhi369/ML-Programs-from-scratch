```python
def k_means(data, k):
    centroids = []

    for i in range(k):
        rand = np.random.randint(0,len(X)-1)
        centroids.append( [ X[rand], Y[rand] ] )
        
    while True:
            
        old_centroids = centroids

        labels = []

        for point in data:
            min_dist = float('inf') # +infinity
            label = None

            for i in range(k):
                new_dist = np.sqrt( (centroids[i][0] - point[0])**2 + (centroids[i][1] - point[1])**2 )
                if new_dist < min_dist:
                    min_dist = new_dist
                    label = i

            labels.append(label)

        new_centroids = [[0,0] for i in range(k)]
        counts = [0]*k

        for point, label in zip(data, labels):
            new_centroids[label][0] += point[0]
            new_centroids[label][1] += point[1]

            counts[label] += 1

        for i in range(k):
            new_centroids[i][0] = new_centroids[i][0]/counts[i]
            new_centroids[i][1] = new_centroids[i][1]/counts[i]
        
        if should_stop(old_centroids, new_centroids, k):
            break

    return labels, new_centroids
        
```
