# NOW we implement the model

```python
def Logistic(x, y, w, b, alpha, iters):
    
    m,n = x.shape
    
    for iter in range(iters):
        
        dj_dw = np.zeros(n)
        dj_db = 0
        
        for i in range(m):
            z = np.dot(x[i], w) + b
            y_pred = 1/(1 + np.exp(-z))
            
            for j in range(n):
                dj_dw[j] = dj_dw[j] + (y_pred - y[i])*x[i][j]
            
            dj_db = dj_db + (y_pred - y[i])
        
        dj_dw = dj_dw/m
        dj_db = dj_db/m
        
        w = w - alpha*dj_dw
        b = b - alpha*dj_db
    
    return w,b
```

```python
alpha = 0.01
iters = 1000
w = np.zeros(x.shape[1])
b = 0

w,b = Logistic(x, y, w, b, alpha, iters)
print(f"features w: {w} and b:{b}")
```

```python
def predict(x, w, b):
    y = []
    for i in range(x.shape[0]):
        z = np.dot(x[i], w) + b
        y_pred = 1/(1 + np.exp(-z))
        y.append(y_pred)
    return y
```

```python
x = np.array([[0.65, 0.44, 0.89, 0.63],
     [0.39, 0.87, 0.46, 0.33],
     [0.70, 0.45, 0.33, 0.70]])


y = predict(x, w, b)
```
